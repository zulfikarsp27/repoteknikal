```python
# Import Packages
import sqlite3
import pandas as pd
import os
import shutil
```

```python
# Setup
if os.path.exists("data_penjualan.db"):
    os.remove("data_penjualan.db")
else:
    print("The file does not exist")

shutil.copyfile("data_penjualan_ori.db", "data_penjualan.db")
```

## Jawaban Soal 1
Tampilkan data order yg mempunyai order amount dibawah 1000

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Membuat Query SQL
result  = connection.cursor()
command = """
             SELECT * FROM orders WHERE ORD_AMOUNT < '1000.00';
          """

# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```

## Jawaban Soal 2
Tampilkan informasi agent yang bernama 'Ivan'.

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Membuat Query SQL
result  = connection.cursor()
command = """
             SELECT * FROM agents WHERE AGENT_NAME = 'Ivan';

          """
# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```

## Jawaban Soal 3
Tampilkan data agent yg bekerja di Bangalore, London, dan Torento.

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Membuat Query SQL
result  = connection.cursor()
command = """
           SELECT * FROM agents WHERE WORKING_AREA IN ('Bangalore','London','Torento');
          """

# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```

## Jawaban Soal 4
Tampilkan informasi customer yang mempunyai payment amount diatas 8000 dan urutkan berdasarkan nama customer dari A sampai Z

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Membuat Query SQL
result  = connection.cursor()
command = """
            SELECT * FROM customer WHERE PAYMENT_AMT > '8000.00' ORDER BY CUST_NAME ASC;
          """

# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```

## Jawaban Soal 5
Tampilkan data customer yang tidak tinggal di USA dan memiliki grade lebih dari 1.

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Membuat Query SQL
result  = connection.cursor()
command = """
            SELECT * FROM customer WHERE CUST_COUNTRY NOT IN ('USA') AND GRADE > '1';
          """

# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```
